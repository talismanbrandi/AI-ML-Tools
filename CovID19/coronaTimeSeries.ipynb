```python
import pandas as pd
import tensorflow as tf
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Read the data
data_confirmed_raw = pd.read_csv('data/time_series_covid19_confirmed_global.csv').drop(['Province/State', 'Lat', 'Long'], axis=1)
data_deaths_raw = pd.read_csv('data/time_series_covid19_deaths_global.csv').drop(['Province/State', 'Lat', 'Long'], axis=1)
data_recovered_raw = pd.read_csv('data/time_series_covid19_recovered_global.csv').drop(['Province/State', 'Lat', 'Long'], axis=1)

data_confirmed = data_confirmed_raw.groupby('Country/Region', as_index=False).sum()
data_confirmed = data_confirmed.sort_values(by=data_confirmed.keys()[-1], ascending=False, ignore_index = True).T
data_deaths = data_deaths_raw.groupby('Country/Region', as_index=False).sum()
data_deaths = data_deaths.sort_values(by=data_deaths.keys()[-1], ascending=False, ignore_index = True).T
data_recovered = data_recovered_raw.groupby('Country/Region', as_index=False).sum()
data_recovered = data_recovered.sort_values(by=data_recovered.keys()[-1], ascending=False, ignore_index = True).T

data_confirmed.columns = data_confirmed.iloc[0]
data_deaths.columns = data_deaths.iloc[0]
data_recovered.columns = data_recovered.iloc[0]

data_confirmed = (data_confirmed.iloc[1:])
data_deaths = data_deaths.iloc[1:]
data_recovered = data_recovered.iloc[1:]
```

```python
data_confirmed.plot(y=data_confirmed.columns[:10], figsize=(10,10), grid=True, logy=True, linewidth=3, alpha=0.75, legend=False)
plt.xlabel('Date', fontsize=22)
plt.ylabel('Number of Confirmed Cases', fontsize=22)
plt.title('Time Series of CovID-19 Cases till 3/27/2020', fontsize=22)
plt.tick_params(labelsize=16)
plt.figlegend(fontsize=12, loc=4, bbox_to_anchor=(0.9, 0.125))
plt.savefig('time.png')
plt.show()
```
