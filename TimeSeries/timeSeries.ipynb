```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data/monthly-sunspots.txt", index_col=0)
data.plot()
```

```python
X = data.values
train_size = int(len(X) * 0.8)
train, test = X[:train_size], X[train_size:]
plt.plot(train)
plt.plot([None for i in train] + [x for x in test])
```
